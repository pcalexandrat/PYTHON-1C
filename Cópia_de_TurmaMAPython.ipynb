```python
import random
dicas_de_logica = [
    "Organize o passo a passo em pseudoc√≥digo antes de escrever o c√≥digo",
    "Anote dados importantes para explicar",
    "Digite a programa√ß√£o e se atente a indenta√ß√£o",
    "Teste seu programa com diferentes valores",
]

while True:
  pedido = input("Digite 'dica' para receber uma dica ou 'sair' para encerrar ").lower()

  if pedido == "dica":
    dica = random.choice(dicas_de_logica) # escolhe uma dica aleat√≥riasair

    print(f"Dica: {dica}")
  elif pedido == "sair":
    print("At√© a pr√≥xima!")
    break
  else:
    print("Comando inv√°lido. Tente novamente.")
```
